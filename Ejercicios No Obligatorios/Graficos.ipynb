```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1024 #numero de puntos
np.random.seed(32) #semilla aleatoria
x = np.random.normal(0,1,n) 
y = np.random.normal(0,1,n)
```

# Imagen 1

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Imagen 2

```python
plt.figure(figsize=(6,6))
plt.scatter(x,y, c= y, cmap= 'RdYlGn')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Imagen 3

```python
size = []

for i in range(0,len(y)):
    if i < 0:
        size.append(1)
    if i >= 0:
        size.append(2)

plt.figure(figsize=(6,6))
plt.scatter(x,y, s= size)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
